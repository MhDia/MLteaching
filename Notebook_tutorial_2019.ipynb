```python
A = np.array([[1,2,3],[4,5,6]]) # 2-d array (i.e. matrix) defined using a list of lists
print("A = ",A)
```

```python
Z = np.zeros((2,2)) # 2 by 2 matrix defined using the zeros built-in method
O = np.ones((3,3)) # 2 by 2 matrix defined using the ones built-in method
print("Z = ",Z)
print("O = ",O)
```

> **Note:** In ``zeros`` and ``ones`` methods with more than one dimension (e.g. matrix), the dimensions are passed as a tuple unlike the random.randn method. For example: np.zeros(2,2) will throw a type error, while np.zeros((2,2)) will work.
You can check the complete documentation of each method using ``shift+tab``.

#### Special case of NumPy rank-1 arrays

```python
np.random.seed(0)
a = np.random.randn(3)
print("a = ",a)
print("shape of a =",a.shape)
```

Note that the shape is ``(3,)`` not ``(3,1)`` or ``(1,3)``. This an example of a rank-1 array in python where the vector is neither a row-vector nor a column vector (it can be both). This inconsistency might cause some common bugs when writing a code for machine learning applications (especially in broadcasting). To avoid such behavior, it is recommended to specify both dimensions or use the ``reshape`` method when needed.

```python
np.random.seed(0)
a = np.random.randn(3,1) # column vector
print("a = ",a)
print("shape of a =",a.shape)
```

```python
np.random.seed(0)
a = np.random.randn(1,3) # row vector
print("a = ",a)
print("shape of a =",a.shape)
```

#### Basic 2D line plot

```python
x = np.arange(0,5,0.1)
y = np.exp(x) # the exponential function
plt.plot(x, y, 'b', label="e(x)") # 'b' indicates the color blue
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of exponential function')
plt.legend()
```

In addition to this simple functional approach for plotting, Matplotlib supports an object-oriented approach where a figure is instantiated and then different methods are called off.

```python
# Create empty figure
fig = plt.figure()

# Add axes
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # fractions (0 to 1) of the figure: left, bottom, width, height

# Plot on axes
axes.plot(x, y, 'r', label="e(x)") # 'r' indicates the color red
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Graph of exponential function - OO approach')
axes.legend()
```

More details on line plots [here](https://matplotlib.org/tutorials/introductory/pyplot.html#sphx-glr-tutorials-introductory-pyplot-py). You can also check various types of plots (e.g. bar plots, histograms,etc...) on the [official matplotlib webpage](https://matplotlib.org/gallery/index.html).

#### 3D plot

```python
# imports
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter

# create data
X = np.arange(-5, 5, 0.1)
Y = np.arange(-5, 5, 0.1)
X,Y = np.meshgrid(X, Y)
Z = (X**2 + Y**2)/5

# plot the data
fig = plt.figure(figsize=(20,10))
ax = fig.gca(projection='3d')
ax.view_init(30,600) # the view angle
surf = ax.plot_surface(X, Y, Z, cmap=cm.viridis)


ax.set_xlabel('$x$', fontsize=16)
ax.set_ylabel('$y$', fontsize=16)
ax.set_zlabel(r'$z = (x^2+y^2)/5$', fontsize=16)

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=10)
```

#### Tangent Plane

```python
# create data
X = np.arange(-3, 3, 0.1)
Y = np.arange(-3, 3, 0.1)
X,Y = np.meshgrid(X, Y)
Z = 5.0-2*X**2-Y**2

# plot the data
fig = plt.figure(figsize=(20,10))
ax = fig.gca(projection='3d')
ax.view_init(30,60)
surf = ax.plot_surface(X, Y, Z, cmap=cm.viridis)

ax = fig.gca(projection='3d')
ax.view_init(30,60)

# tangent plane at the point (-1,-1)
X0=-1.0
Y0=-1.0
TX = np.arange(-1.5, 1.5, 0.1)+X0
TY = np.arange(-1.5, 1.5, 0.1)+Y0
TX,TY = np.meshgrid(TX, TY)
Z0 = 5.0-2*X0**2 - Y0**2
TZ = Z0+(-4*X0*(TX-X0)-2*Y0*(TY-Y0))

# plot tangent plane
surf = ax.plot_surface(TX, TY, TZ, cmap=cm.cividis)

ax.set_zlim(-15.0, 5.0)

ax.set_xlabel('$x$', fontsize=12)
ax.set_ylabel('$y$', fontsize=12)
ax.set_zlabel(r'$5-2x^2-y^2$', fontsize=12)
ax.view_init(30, -30)
```

```python
'''
We will now plot the tangent plane at the optimal point,
it will be a horizontal plane parallel to the x-y plane.
'''
# plot the data
fig = plt.figure(figsize=(20,10))
ax = fig.gca(projection='3d')
ax.view_init(30,60)
surf = ax.plot_surface(X, Y, Z, cmap=cm.viridis)

ax = fig.gca(projection='3d')
ax.view_init(30,60)

# tangent plane at the point (0,0)
X0=0
Y0=0
TX = np.arange(-1.5, 1.5, 0.1)+X0
TY = np.arange(-1.5, 1.5, 0.1)+Y0
TX,TY = np.meshgrid(TX, TY)
Z0 = 5.0-2*X0**2 - Y0**2
TZ = Z0+(-4*X0*(TX-X0)-2*Y0*(TY-Y0))

# plot tangent plane
surf = ax.plot_surface(TX, TY, TZ, cmap=cm.cividis)

ax.set_zlim(-15.0, 5.0)

ax.set_xlabel('$x$', fontsize=12)
ax.set_ylabel('$y$', fontsize=12)
ax.set_zlabel(r'$5-2x^2-y^2$', fontsize=12)
ax.view_init(30, -30)
```

#### Discontinuous functions

```python
# create data
X = np.arange(-5, 5, 0.1)
Y = np.arange(-5, 5, 0.1)
X,Y = np.meshgrid(X, Y)
Z = 2*X*Y/(X**2 + Y**2)

# plot the data
fig = plt.figure(figsize=(20,10))
ax = fig.gca(projection='3d')
ax.view_init(25,60)
surf = ax.plot_surface(X, Y, Z, cmap=cm.viridis)

ax.set_zlim(-2, 2)

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=10)
```

```python
delta = 0.01
X = np.arange(-5, 5, delta)
Y = np.arange(-5, 5, delta)
X,Y = np.meshgrid(X, Y)
Z = 2*X*Y/(X**2 + Y**2)

plt.figure()
colors=('green', 'red', 'blue')
V=[-0.99999, 0.0, 0.99999]
CS = plt.contour(X, Y, Z, V, colors=colors)
plt.clabel(CS, inline=1, fontsize=10)
```

Notice the discontinuity at the origin. The function assumes $0$ value along the $x$-axis and $y$-axis, however it has a value $1$ along the diagonals $x=y$ and $x=-y$.
